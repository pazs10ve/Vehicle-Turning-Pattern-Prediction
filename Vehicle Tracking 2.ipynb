```python
import cv2
import numpy as np
import pandas as pd
from ultralytics import YOLO
from collections import defaultdict
import threading
from queue import Queue
from tqdm import tqdm
import time
import csv
```

```python
model_path = r"C:\Users\sharm\Python Projects\ML Projects\Bengaluru Mobility Challenge\Bengaluru Mobility Challenge Solution\runs\detect\train\weights\best.pt"
yolo_model = YOLO(model_path)
print("Model loaded successfully!")
```

```python
def process_video(video_path, yolo_model):
    cap = cv2.VideoCapture(video_path)    
    width = int(cap.get(cv2.CAP_PROP_FRAME_WIDTH))
    height = int(cap.get(cv2.CAP_PROP_FRAME_HEIGHT))
    fps = int(cap.get(cv2.CAP_PROP_FPS))
    frame_count = int(cap.get(cv2.CAP_PROP_FRAME_COUNT))

    print(frame_count, fps, width, height)


    tracks = defaultdict(list)    
    
    
    for i in tqdm(range(frame_count), desc="Processing video"):
        success, frame = cap.read()

        if success:
            results = yolo_model.track(frame, persist=True, tracker="bytetrack.yaml")
            annotated_frame = results[0].plot()

            for result in results[0].boxes:
                obj_id = int(result.id)
                bbox = result.xyxy 
                tracks[obj_id].append((i, bbox.cpu().numpy().tolist()))

            #cv2.namedWindow("Vehicle Tracking", cv2.WINDOW_NORMAL)
            #cv2.imshow("Vehicle Tracking", annotated_frame)
            
        else:
            break

    cap.release()
    cv2.destroyAllWindows()

    track_df = pd.DataFrame([(obj_id, frame_id, bbox) for obj_id, frames in tracks.items() for frame_id, bbox in frames],
                            columns=["Object ID", "Frame ID", "Bounding Box"])
    
    return track_df, tracks
```

```python
video_path = r"C:\Users\sharm\Python Projects\ML Projects\Bengaluru Mobility Challenge\Bengaluru Mobility Challenge Solution\Video File\Stn_HD_1_time_2024-05-29T07_30_02_002.mp4"
output_path = r"C:\Users\sharm\Python Projects\ML Projects\Bengaluru Mobility Challenge"
track_df, tracks = process_video(video_path, yolo_model)
```
