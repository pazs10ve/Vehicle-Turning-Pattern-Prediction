```python
def download_file(s3_uri):
  bucket_name, key = s3_uri.replace('s3://', '').split('/', 1)
  file_name = key.split('/')[-1]
  local_file_path = os.path.join(directory, file_name)

  s3_client.download_file(bucket_name, key, local_file_path)
  print(f'Downloaded {file_name}')
```

```python
output_directory = 'Indian Vehicle Dataset'
if not os.path.exists(output_directory):
    os.makedirs(output_directory)
```

```python
def extract_frames(video_path, output_dir, fps=1):
    video_name = os.path.basename(video_path).split('.')[0]
    cap = cv2.VideoCapture(video_path)
    count = 0
    frame_number = 0

    while cap.isOpened():
        ret, frame = cap.read()
        if not ret:
            break

        if count % int(cap.get(cv2.CAP_PROP_FPS)) == 0:
            frame_name = f"{video_name}_frame_{frame_number}.jpg"
            frame_path = os.path.join(output_dir, frame_name)
            cv2.imwrite(frame_path, frame)
            frame_number += 1
            print(frame_number)

        count += 1

    cap.release()
```

```python
video_directory = '/content/Stn_HD_1'

for video_file in os.listdir(video_directory):
  video_path = os.path.join(video_directory, video_file)
  extract_frames(video_path, output_directory)

print('Frames extracted and saved.')
```
