```python
import pandas as pd

results_file = '/content/runs/detect/train/results.csv'
df = pd.read_csv(results_file)

print("Available columns:")
print(df.columns)

print(df.head())
```

```python
df.columns = ['epoch', 'train/box_loss',
       'train/cls_loss', 'train/dfl_loss',
       'metrics/precision(B)', 'metrics/recall(B)',
       'metrics/mAP50(B)', 'metrics/mAP50-95(B)',
       'val/box_loss', 'val/cls_loss',
       'val/dfl_loss', 'lr/pg0',
       'lr/pg1', 'lr/pg2']
df.columns
```

```python
import matplotlib.pyplot as plt
# Create a figure with subplots
fig, (ax1, ax2, ax3) = plt.subplots(3, 1, figsize=(10, 15))

# Plot mAP
ax1.plot(df['epoch'], df['metrics/mAP50-95(B)'], label='mAP50-95')
ax1.set_xlabel('Epoch')
ax1.set_ylabel('mAP')
ax1.set_title('Mean Average Precision (mAP50-95)')
ax1.legend()

ax2.plot(df['epoch'], df['metrics/precision(B)'], label='Precision')
ax2.set_xlabel('Epoch')
ax2.set_ylabel('Precision')
ax2.set_title('Precision')
ax2.legend()

# Plot Recall
ax3.plot(df['epoch'], df['metrics/recall(B)'], label='Recall')
ax3.set_xlabel('Epoch')
ax3.set_ylabel('Recall')
ax3.set_title('Recall')
ax3.legend()

plt.tight_layout()
plt.show()

plt.savefig('/content/training_metrics.png')
print("Training metrics visualization saved as 'training_metrics.png'")
```

```python
sample = model('/content/Stn_HD_1_time_2024-05-15T07_30_02_000_frame_309.jpg')
```

```python
import cv2
import matplotlib.pyplot as plt

img = cv2.imread('/content/Stn_HD_1_time_2024-05-15T07_30_02_000_frame_309.jpg')

sample_results = sample[0]

for result in sample_results.boxes:
    x1, y1, x2, y2 = map(int, result.xyxy[0])
    conf = result.conf[0]
    cls = int(result.cls[0])
    label = f'{model.names[cls]}: {conf:.2f}'
    cv2.rectangle(img, (x1, y1), (x2, y2), (0, 255, 0), 2)
    cv2.putText(img, label, (x1, y1 - 10), cv2.FONT_HERSHEY_SIMPLEX, 0.9, (0, 255, 0), 2)

img = cv2.cvtColor(img, cv2.COLOR_BGR2RGB)

plt.figure(figsize=(12, 8))
plt.imshow(img)
plt.axis('off')
plt.show()
```
